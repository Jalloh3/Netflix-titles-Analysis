## **Total Session Duration Analysis**

```python
df["TotalDuration"] = (
    df["Administrative_Duration"] +
    df["Informational_Duration"] +
    df["ProductRelated_Duration"]
)

df.boxplot(column="TotalDuration", by="Revenue")
plt.title("Total Session Duration by Purchase Status")
plt.suptitle("")
plt.xlabel("Revenue")
plt.ylabel("Total Duration (seconds)")
plt.show()
```
